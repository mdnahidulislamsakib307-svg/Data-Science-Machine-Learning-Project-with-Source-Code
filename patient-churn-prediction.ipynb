```python
df = pd.read_csv("/kaggle/input/patient-churn-prediction-dataset-for-healthcare/patient_churn_dataset.csv")
```

```python
df.head()
```

```python
x = df.drop(['Churned'],axis=1)
y = df['Churned']
```

```python
numerical_cols = x.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
categorical_cols = x.select_dtypes(include=['object']).columns.tolist()
```

```python
from sklearn.model_selection import train_test_split
from sklearn .compose import ColumnTransformer
from sklearn .impute import SimpleImputer
from sklearn .preprocessing import OneHotEncoder,StandardScaler
from sklearn .pipeline import Pipeline
```

```python
numerical_transformer = Pipeline (steps=[
                                 ('imputer',SimpleImputer(strategy = 'mean')),
                                ('scaler',StandardScaler())
])
```

```python
categorical_transformer = Pipeline(steps=[
                                    ('imputer',SimpleImputer(strategy='most_frequent')),
                                    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor =  ColumnTransformer(transformers=[
                                 ('num',numerical_transformer,numerical_cols),
                                 ('cat',categorical_transformer,categorical_cols)
])
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_1 = Pipeline(steps=[
                      ('pre',preprocessor),('reg',RandomForestRegressor(n_estimators=100,random_state=42))
])
```

```python
model_1.fit(X_train,y_train)
```

```python
from sklearn .metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
randomforest_prediction= model_1.predict(X_test)

print(f'Accuracy:{r2_score(randomforest_prediction,y_test):.4f}')
```

```python
from sklearn .tree import DecisionTreeRegressor 
```

```python
model_2 = Pipeline(steps=[
             ('pre',preprocessor ),('reg',DecisionTreeRegressor(max_depth=7,random_state=42))
])
```

```python
model_2.fit(X_train,y_train)
```

```python
dicisiontree_prediction = model_2.predict(X_test)


print(f'Accuracy:{r2_score(dicisiontree_prediction,y_test):.4f}')
```

```python
from sklearn .linear_model import LogisticRegression
```

```python
model_3 = Pipeline(steps=[
                       ('pre',preprocessor),('reg',LogisticRegression())
])
```

```python
model_3.fit(X_train,y_train)
```

```python
logistic_prediction = model_3.predict(X_test)



print(f'Accuracy:{r2_score(logistic_prediction,y_test):.4f}')
```
